```
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sb
import numpy as np
```

```
df=pd.read_csv('salary.csv')

# it will provide inforamation about the file salary.csv
df.info()
```

```
# top 5 by default when you provide no arrgument
df.head()
```

```
sb.countplot(df['YearsExperience'])
```

```
# data with x axis 
x=df.iloc[:,0].values
x.shape
```

```
y=df.iloc[:,-1].values
```

```
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.2,random_state=0)
```

```
from sklearn.linear_model import LinearRegression
regr=LinearRegression()
```

```
trained=regr.fit(train_x.reshape(-1,1),train_y)
```

```
#predict modal
output=trained.predict(test_x.reshape(-1,1))
```

```
#plotiing data we have
plt.scatter(train_x,train_y,label='exp. Vs Sal',color='Red')
plt.plot(train_x,trained.predict(train_x.reshape(-1,1)),color="green")

#maiking modal of untrained data that is testing data 
#plt.plot(test_x,output,color="blue")

plt.xlabel('Experience')
plt.ylabel('salary')
plt.legend()
```
