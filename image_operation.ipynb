```
import cv2 #import library opencv
import numpy as np   #import numpy in np variable
import matplotlib.pylab as plt
from google.colab.patches import cv2_imshow  #colab 
```

```
green=cv2.imread('i.jpg')      # image read using opencv
```

```
for i in range(800):
  for j in range(550):
      green[i][j][0]=0
      green[i][j][0]=128
      green[i][j][0]=0
cv2_imshow(green)
```

```
image = cv2.imread("i.jpg")
resized=image
#ORIGNAL IMAGE PRINT
cv2_imshow(image)

ones=np.zeros([800,550])

#r = 150.0 / image.shape[1]
#dim = (150, int(image.shape[0] * r))
#resized = cv2.resize(image, dim, interpolation=cv2.INTER_AREA)
#cv2_imshow(resized)

# lower_white & upper white is the range of color raplcement
lower_white = np.array([148, 144, 143], dtype=np.uint8)
upper_white = np.array([186, 182, 183], dtype=np.uint8)

# create a 2-bit color mask over the image
mask = cv2.inRange( resized, lower_white, upper_white) 
#PRINTING IMAGE MASK
cv2_imshow(mask)

res = cv2.bitwise_not(resized, resized, mask)

cv2_imshow(res) # gives black background
```
