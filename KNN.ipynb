```
from sklearn.datasets import load_iris
```

```
iris=load_iris()
features=iris.data
label=iris.target
```

```
from sklearn.model_selection import train_test_split
train_ir,test_ir,train_lb,test_lb=train_test_split(features,label)  #by default 25% (in testing)
```

```
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
```

```
clf=KNeighborsClassifier(n_neighbors=5)
clf1=DecisionTreeClassifier()
```

```
#traininggdataset

knn_trained=clf.fit(train_ir,train_lb)
dt_trained=clf1.fit(train_ir,train_lb)
```

```
#prediction
predicted=knn_trained.predict(test_ir)
```

```
from sklearn.metrics import accuracy_score
knn_acc=accuracy_score(test_lb,predicted)
knn_acc
```
